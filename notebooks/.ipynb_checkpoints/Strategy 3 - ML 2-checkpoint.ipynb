```python
from strategies.ANN_returns import ANNReturnStrategy
from strategies.LSTM import LSTMStrategy
from strategies.CNN import CNNStrategy
from strategies.ANN_prices import ANNPriceStrategy
from strategies.LSTM_multiperiod import LSTMMultiPeriodStrategy
from strategies.CNN_LSTM import HybridCNNLSTMStrategy
from utils.backtest import Backtest
from utils.data_handler import DataHandler
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data_handler = DataHandler(ticker='MSFT', start_date='2020-01-01', end_date='2024-01-01')
data = data_handler.fetch_data()
```

### Technicals in a simple NN to do a single-period prediction 

```python
# Prices
price_strategy = ANNPriceStrategy()
backtest_prices = Backtest(data, price_strategy)
results_prices = backtest_prices.run()
print(results_prices)
```

```python
# Returns
returns_strategy = ANNReturnStrategy()
backtest_returns = Backtest(data, returns_strategy)
results_returns = backtest_returns.run()
print(results_returns)
```

#### Price prediction vs Return prediction

```python
# 1. Stratégie Prix
price_strategy.generate_signals(data)  # Ceci entraîne le modèle

# Récupérer les prédictions de prix
test_size = int(len(data) * 0.2)
real_prices = data['Close'].iloc[-test_size:].values
predicted_prices = price_strategy.scaler_y.inverse_transform(
    price_strategy.model.predict(price_strategy.scaler_X.transform(
        price_strategy.compute_indicators(data)[['SMA20', 'SMA50', 'Price_SMA20', 'RSI', 'MACD',
                                               'BB_upper', 'BB_lower', 'Price_BB_up']].iloc[-test_size:]
    ))
)

# 2. Stratégie Returns
returns_strategy.generate_signals(data)  # Ceci entraîne le modèle

# Récupérer les prédictions de returns
real_returns = data['Close'].pct_change().iloc[-test_size:].values
predicted_returns = returns_strategy.model.predict(
    returns_strategy.scaler.transform(
        returns_strategy.compute_indicators(data)[['RSI', 'RSI_Change', 'MACD', 'SMA_cross', 'BB_width']].iloc[-test_size:]
    )
).flatten()

# Visualisation
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))

# Prix
ax1.plot(real_prices, label='Prix Réels', color='blue')
ax1.plot(predicted_prices, label='Prix Prédits', color='red', linestyle='--')
ax1.set_title('Comparaison Prix Réels vs Prédits')
ax1.legend()
ax1.grid(True)

# Returns
ax2.plot(real_returns, label='Returns Réels', color='blue')
ax2.plot(predicted_returns, label='Returns Prédits', color='red', linestyle='--')
ax2.set_title('Comparaison Returns Réels vs Prédits')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# Métriques d'erreur
print("\nMétriques pour la Prédiction des Prix:")
print(f"MSE: {mean_squared_error(real_prices, predicted_prices):.2f}")
print(f"R²: {r2_score(real_prices, predicted_prices):.2f}")

print("\nMétriques pour la Prédiction des Returns:")
print(f"MSE: {mean_squared_error(real_returns, predicted_returns):.6f}")
print(f"R²: {r2_score(real_returns, predicted_returns):.2f}")
```

### LSTM

```python
my_strategy = LSTMStrategy()
backtest = Backtest(data, my_strategy)
results = backtest.run()
# my_strategy.plot_signals(data)
print(results)
```

```python
# Pour prédictions multi-périodes avec LSTM
lstm_multi = LSTMMultiPeriodStrategy(forecast_horizon=5)  # Prédiction sur 5 périodes
backtest_lstm = Backtest(data, lstm_multi)
results_lstm = backtest_lstm.run()
print(results_lstm)
```

```python
# Création des modèles
lstm_price = LSTMMultiPeriodStrategy(predict_returns=False, window_size=20, forecast_horizon=5)
lstm_returns = LSTMMultiPeriodStrategy(predict_returns=True, window_size=20, forecast_horizon=5)

# Backtest des deux stratégies
backtest_price = Backtest(data, lstm_price)
backtest_returns = Backtest(data, lstm_returns)

results_price = backtest_price.run()
results_returns = backtest_returns.run()

# Visualisation des prédictions multi-périodes
def plot_multiperiod_predictions(strategy, title):
    plt.figure(figsize=(15, 6))
    
    # Tracer les prédictions pour les 5 derniers points
    for i in range(5):
        plt.plot(strategy.y_test[-5:, i], 'b-', label=f'Réel T+{i}' if i==0 else '')
        plt.plot(strategy.y_pred[-5:, i], 'r--', label=f'Prédit T+{i}' if i==0 else '')
    
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_multiperiod_predictions(lstm_price, "Prédictions Multi-périodes des Prix")
plot_multiperiod_predictions(lstm_returns, "Prédictions Multi-périodes des Rendements")

# Comparaison des performances
print("\nPerformances de la stratégie Prix:")
print(f"Return: {results_price['Return']:.2%}")
print(f"Sharpe: {results_price['Sharpe Ratio']:.2f}")
print(f"Max Drawdown: {results_price['Max Drawdown']:.2%}")

print("\nPerformances de la stratégie Returns:")
print(f"Return: {results_returns['Return']:.2%}")
print(f"Sharpe: {results_returns['Sharpe Ratio']:.2f}")
print(f"Max Drawdown: {results_returns['Max Drawdown']:.2%}")
```
