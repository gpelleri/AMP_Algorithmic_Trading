# Neural Network Strategy with Technical Indicators

This notebook implements and analyzes a trading strategy that combines technical indicators with neural networks for S&P 500 stocks. We'll examine both price and return predictions using cross-validation.

```python
import sys
import os

# Get the parent directory of the "notebook" folder
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))

# Add it to the Python path
sys.path.append(project_root)
```

```python
from strategies.technical_nn import TechnicalNNStrategy
from utils.backtest import Backtest
from utils.data_handler import DataHandler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

## 1. Data Loading
Let's load some S&P 500 stocks for our analysis

```python
# List of S&P 500 stocks to analyze
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
start_date = '2020-01-01'
end_date = '2024-01-01'

# Store results for comparison
results_price = {}
results_return = {}
```

## 2. Price Prediction Analysis

```python
for ticker in tickers:
    print(f"\nAnalyzing {ticker} - Price Prediction")
    
    # Load data
    data_handler = DataHandler(ticker=ticker, start_date=start_date, end_date=end_date)
    data = data_handler.fetch_data()
    
    # Initialize and run strategy
    strategy = TechnicalNNStrategy(prediction_type='price', n_splits=5, epochs=50)
    backtest = Backtest(data, strategy)
    results = backtest.run()
    
    # Store results
    results_price[ticker] = {
        'backtest_results': results,
        'cv_scores': strategy.cv_scores
    }
    
    # Plot results
    strategy.plot_signals(data)
```

## 3. Return Prediction Analysis

```python
for ticker in ["AAPL"]:
    print(f"\nAnalyzing {ticker} - Return Prediction")
    
    # Load data
    data_handler = DataHandler(ticker=ticker, start_date=start_date, end_date=end_date)
    data = data_handler.fetch_data()
    
    # Initialize and run strategy
    strategy = TechnicalNNStrategy(prediction_type='return', n_splits=5, epochs=50)
    backtest = Backtest(data, strategy)
    results = backtest.run()
    
    # Store results
    results_return[ticker] = {
        'backtest_results': results,
        'cv_scores': strategy.cv_scores
    }
    
    # Plot results
    strategy.plot_signals(data)
```

## 4. Results Comparison

```python
# Compare price vs return prediction performance
comparison_data = []

for ticker in tickers:
    comparison_data.append({
        'Ticker': ticker,
        'Price_Pred_MSE': np.mean(results_price[ticker]['cv_scores']),
        'Price_Pred_Std': np.std(results_price[ticker]['cv_scores']),
        'Return_Pred_MSE': np.mean(results_return[ticker]['cv_scores']),
        'Return_Pred_Std': np.std(results_return[ticker]['cv_scores']),
        'Price_Strategy_Return': results_price[ticker]['backtest_results']['Return'],
        'Return_Strategy_Return': results_return[ticker]['backtest_results']['Return'],
    })

comparison_df = pd.DataFrame(comparison_data)
print("\nStrategy Comparison:")
print(comparison_df)
```

```python
# Visualize comparison
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot MSE comparison
x = np.arange(len(tickers))
width = 0.35

ax1.bar(x - width/2, comparison_df['Price_Pred_MSE'], width, label='Price Prediction')
ax1.bar(x + width/2, comparison_df['Return_Pred_MSE'], width, label='Return Prediction')
ax1.set_ylabel('Mean Squared Error')
ax1.set_title('Prediction Error Comparison')
ax1.set_xticks(x)
ax1.set_xticklabels(tickers)
ax1.legend()

# Plot Strategy Returns comparison
ax2.bar(x - width/2, comparison_df['Price_Strategy_Return'], width, label='Price Strategy')
ax2.bar(x + width/2, comparison_df['Return_Strategy_Return'], width, label='Return Strategy')
ax2.set_ylabel('Strategy Return (%)')
ax2.set_title('Trading Performance Comparison')
ax2.set_xticks(x)
ax2.set_xticklabels(tickers)
ax2.legend()

plt.tight_layout()
plt.show()
```
